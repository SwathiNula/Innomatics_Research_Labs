```python
# Task 1:Login check

saved_user = "admin"
saved_pass = "1234"

user = input("Enter username: ")
pwd = input("Enter password: ")

if user == saved_user and pwd == saved_pass:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2: Pass / Fail analyzer

marks = [45, 78, 90, 33, 60]

passed = 0
failed = 0

for m in marks:
    if m >= 50:
        passed = passed + 1
    else:
        failed = failed + 1

print("Passed:", passed)
print("Failed:", failed)
```

```python
# Task 3: Simple data cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    new_name = ""

    for ch in name:
        # ignoring spaces
        if ch == " ":
            continue

        # converting capital letters to small letters
        if ch >= 'A' and ch <= 'Z':
            new_name = new_name + chr(ord(ch) + 32)
        else:
            new_name = new_name + ch

    cleaned_names.append(new_name)

print("Cleaned names:")
for name in cleaned_names:
    print(name)
```

```python
# Task 4: Message length analyzer

msgs = ["Hi", "Welcome to the platform", "OK"]

for m in msgs:
    c = 0

    for ch in m:
        c = c + 1

    print("Message:", m)
    print("Length:", c)

    if c > 10:
        print("Long message")
    else:
        print("Short message")

    print()
```

```python
# Task 5: Error message detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

cnt = 0

for x in logs:
    if x == "ERROR":
        cnt = cnt + 1

print("Total ERROR count:", cnt)
```
